```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        
    ## Insert at the end of the linked list
    def insertAtEnd(self, new_data):
        new_node = Node(new_data)
        
        ## If the linked list is empty
        if self.head is None:
            self.head = new_node
            return
        
        ## Traverse to the end and insert the new node
        temp = self.head
        while temp.next:
            temp = temp.next
        
        temp.next = new_node
    
    ## Insert after a specific node in the linked list
    def insertAfterNode(self, prev_node, new_data):
        if prev_node is None:
            print("Given node must be available inside the linked list")
            return
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node
    
    ## Print the linked list
    def printList(self):
        temp = self.head
        while temp:
            print(str(temp.data) + " ", end=" ")
            temp = temp.next
        print()

    ## Find the node with a specific value
    def findNode(self, target):
        temp = self.head
        while temp:
            if temp.data == target:
                return temp
            temp = temp.next
        return None


## Driver code
llist = LinkedList()

## Function calling to create the initial linked list
llist.insertAtEnd(10)
llist.insertAtEnd(20)
llist.insertAtEnd(50)
llist.insertAtEnd(5)
llist.insertAtEnd(1)
llist.insertAtEnd(100)

print("Original Linked List:")
llist.printList()

print()
llist.insertAfterNode(llist.head.next, 30)
print("Insertion after any node")
llist.printList()


## Find the node with value 5
prev_node = llist.findNode(5)

## Insert 30 after the node with value 5
if prev_node:
    llist.insertAfterNode(prev_node, 30)
    print("Insertion after node with value 5:")
    llist.printList()
else:
    print("Node with value 5 not found.")
```
