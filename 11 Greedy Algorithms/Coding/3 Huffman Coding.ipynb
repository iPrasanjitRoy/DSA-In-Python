```python
# A Huffman Tree Node
import heapq

class node:
    def __init__(self, freq, symbol, left=None, right=None):
    # FREQUENCY OF SYMBOL 
        self.freq = freq

    # SYMBOL NAME (CHARACTER)
        self.symbol = symbol

    # NODE LEFT OF CURRENT NODE 
        self.left = left # Left child 

    # NODE RIGHT OF CURRENT NODE  
        self.right = right # Right child

    # TREE DIRECTION (0/1)
        self.huff = ''
    
    # COMPARE BASED ON FREQUENCY 
    # NXT IS ANOTHER OBJ OF NODE BEING COMPARED 
    def __lt__(self, nxt):
        return self.freq < nxt.freq



## UTILITY FUNCTION TO PRINT HUFFMAN CODES FOR ALL SYMBOLS IN THE NEWLY CREATED HUFFMAN TREE 
def printNodes(node, val=''):
    
    # HUFFMAN CODE FOR CURRENT NODE 
    newVal = val + str(node.huff) # Append 0 or 1 to form Huffman code

    # IF NODE IS NOT AN LEAF NODE THEN TRAVERSE INSIDE IT 
    if(node.left):
        printNodes(node.left, newVal) # Go left (add '0')
    if(node.right):
        printNodes(node.right, newVal) # Go right (add '1')


    # IF NODE IS LEAF NODE THEN DISPLAY ITS HUFFMAN CODE 
    if(not node.left and not node.right):
        print(f"{node.symbol} -> {newVal}")






## DRIVER CODE 
## CHARACTERS FOR THE HUFFMAN TREE  
chars = ['a', 'b', 'c', 'd', 'e', 'f']

## FREQUENCY OF CHARACTERS 
freq = [5, 9, 12, 13, 16, 45]

## LIST CONTAINING UNUSED NODES 
nodes = [] 


## CONVERTING CHARACTERS AND FREQUENCIES INTO HUFFMAN TREE NODES 
## heapq.heappush(nodes, node(freq[0], chars[0]))  # Push ('a', 5) 
## AT THE END OF THE LOOP, NODES (A MIN-HEAP) CONTAINS: 
## [(5, 'a'), (9, 'b'), (12, 'c'), (13, 'd'), (16, 'e'), (45, 'f')] 

for x in range(len(chars)):
    heapq.heappush(nodes, node(freq[x], chars[x]))



while len(nodes) > 1:

    ## SORT ALL THE NODES IN ASCENDING ORDER BASED ON THEIR FREQUENCY     
    left = heapq.heappop(nodes) 
    right = heapq.heappop(nodes) 

    ## ASSIGN VALUE TO THESE NODES IN THE HUFFMAN TREE
    left.huff = 0 
    right.huff = 1 

    ## COMBINE THE 2 SMALLEST NODES TO CREATE NEW NODE AS THEIR PARENT 
    ## Combine 'a' (5) + 'b' (9) → New Node (14, 'ab') 
    ## Combine 'c' (12) + 'd' (13) → New Node (25, 'cd') 
    ## Combine 'ab' (14) + 'e' (16) → New Node (30, 'abe') 
    ## Combine 'f' (45) + 'cdaeb' (55) → Root Node (100, 'fcdaeb') 
    newNode = node(left.freq+right.freq, left.symbol+right.symbol, left, right) 

    heapq.heappush(nodes, newNode)


## HUFFMAN TREE IS READY! 
## ROOT ONLY REMAINING NODE
printNodes(nodes[0])   
```

```python
# A Huffman Tree Node
import heapq

class node:
    def __init__(self, freq, symbol, left=None, right=None):
        self.freq = freq
        self.symbol = symbol
        self.left = left # Left child 
        self.right = right # Right child
        self.huff = ''
    

    def __lt__(self, nxt):
        return self.freq < nxt.freq



def printNodes(node, val='', is_root=True):
    newVal = val + str(node.huff)

    # Check If It's The Root Node 
    if is_root:
        print(f"Root Node '{node.symbol}' -> {newVal}")
    elif node.left or node.right:
        print(f"Internal Node '{node.symbol}' -> {newVal}")


    if node.left:
        printNodes(node.left, newVal, is_root=False)
    if node.right:
        printNodes(node.right, newVal, is_root=False)


    if not node.left and not node.right:
        print(f"Leaf Node '{node.symbol}' -> {newVal}")




chars = ['a', 'b', 'c', 'd', 'e', 'f']
freq = [5, 9, 12, 13, 16, 45]

nodes = [] 

for x in range(len(chars)):
    heapq.heappush(nodes, node(freq[x], chars[x]))


while len(nodes) > 1:
    left = heapq.heappop(nodes) 
    right = heapq.heappop(nodes) 

    left.huff = 0 
    right.huff = 1 

    newNode = node(left.freq+right.freq, left.symbol+right.symbol, left, right) 

    heapq.heappush(nodes, newNode)



printNodes(nodes[0])    
```
